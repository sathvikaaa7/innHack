```python
# Problem Statement 1: Employee Performance Bonus Eligibility

employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
x = max(employees.values())

li = []
for name, score in employees.items():
    if score == x:
        li.append(name)


names = ", ".join(li)

print(f"Top Performers Eligible for Bonus: {names} (Score: {x})")


```

```python
# Problem Statement 2: Search Query Keyword Analysis

s = "Buy mobile phone buy phone online"
s = s.lower()

for ch in ",.!?":
    s = s.replace(ch, "")

words = s.split()
keyC = {}

for i in words:
    if words.count(i) > 1:
        keyC[i] = words.count(i)

print(keyC)

```

```python
# Problem Statement 3: Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
result = []
for i,v  in enumerate(sensor_readings):
  if v%2 == 0:
    result.append((i,v))
print(result)
```

```python
# Problem Statement 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

d = []

for e in emails:
    parts = e.split("@")
    d.append(parts[1])

print(d)
gc = 0
yc = 0

for i in d:
  if i == "gmail.com":
    gc += 1
  else:
    yc += 1

gp = (gc / len(emails)) * 100
yp = (yc / len(emails)) * 100

print(f"gmail.com: {gp}%")
print(f"yahoo.com: {yp} %")


```

```python
# Problem Statement 5: Sales Spike Detection
import math

sales = [1200, 1500, 900, 2200, 1400, 3000]
avg = sum(sales) / len(sales)

for i,v  in enumerate(sales):
  if v >= avg:
    print(f"Day {i+1}: {v}")
```

```python
# Problem Statement 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

d1 = {}

for i in user_ids:
   d1[i] = d1.get(i, 0) + 1

for user, count in d1.items():
    if count > 1:
        print(f"{user} → {count} times")
```
