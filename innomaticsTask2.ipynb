```python
#2 Highest Salary from Employee Data
employeeDict = {"Ravi":70000, "Lekha":77000, "Rupa":58000}
max = 0

for key, value in employeeDict.items():
  if value > max:
    max = value
    emp = key
print(f"Highest Salary:{emp}:{max}")
```

```python
#3. Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]

minimum = numbers[0]
maximum = numbers[0]

for num in numbers:
    if num < minimum:
        minimum = num
    if num > maximum:
        maximum = num

print("List:",numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
#4. Count Products Above a Price Threshold

prices = [450, 1200, 899, 1500, 300]
count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)

```

```python
#5.Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

present_count = 0

for status in attendance:
    if status == "P":
        present_count += 1

total_days = len(attendance)
attendance_percentage = (present_count / total_days) * 100

print("Attendance Percentage:", attendance_percentage)
```

```python
#6. Remove Duplicate Phone Numbers

phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)

```

```python
#7. Count Character Frequency

text = "pythonpp"
char_count = {}

for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

print(char_count)
```
