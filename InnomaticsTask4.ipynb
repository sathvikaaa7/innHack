```python
# Problem 1: Mobile Recharge Validation System

def validate(n):
  li = [199, 299, 399, 599]
  while True:
    if n in li:
      print("Recharge Plan Valid")
      break
    else:
      n = int(input("Enter Plan Amount Again"))

n = int(input("Enter Plan Amount"))
validate(n)
```

```python
# Problem 2: Inventory Reorder Alert System

def stockAlert(d1):
  for k,v in d1.items():
    if v < 15:
      print(f"{k}: Reorder Alert")
    else:
      print(f"{k}: Stock Ok")

d1 = {"Mobile":35, "HeadPhone":13, "Keyboard":22, "Charger":7}
stockAlert(d1)
```

```python
# Problem 3: Student Result Processing System

def studentResult(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


marks = [60, 55, 70, 45]
result = studentResult(marks)
print(result)

```

```python
# Problem 4: Cab Fare Estimator with Retry Option

def fare(distance, hour):
    base_fare = 50
    kmPlus = 12

    price = base_fare + (distance * kmPlus)
    peak_hours = [6,7,8]
    if hour in peak_hours:
        price += price * 0.25

    return price


while True:
    distance = float(input("Enter distance in km: "))
    hour = int(input("Booking Time? (hour)"))

    total_fare = fare(distance, hour)
    print(f"Total Cab Fare: ₹{total_fare:.2f}")

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

```python
# Problem 5: Employee Attendance Eligibility

def check(l1):
  count_P = 0
  for i in l1:
    if i == 'P':
      count_P += 1
  percentage = (count_P/len(l1)) * 100
  if percentage >= 75:
    return "Eligible"
  else:
    return "Not Eligible"


employee1_attendance = ['P', 'A', 'P', 'A', 'P', 'P', 'P', 'P','P','P','P',]
employee2_attendance = ['P', 'A', 'P', 'A', 'P', 'P', 'P', 'A','P','P','P',]
print(f"Employee 1 is {check(employee1_attendance)}")
print(f"Employee 2 is {check(employee2_attendance)}")
```

```python
# Problem 6: Password Strength Checker

def check(password1):

  if len(password1) < 8:
    return "Weak Password"

  has_digit = False
  has_spl_char = False
  spl_chars = "@#$"

  for i in password1:
    if i.isdigit():
      has_digit = True
    if i in spl_chars:
      has_spl_char = True

  if has_digit and has_spl_char:
    return "Strong Password"
  else:
    return "Weak Password"


p1 = input("Enter password: ")
print(check(p1))
```
